```
df=pd.read_csv('/content/gdrive/My Drive/CSC 522 ALDA/Project/genre_split.csv',header=0,encoding='latin-1')

df_awards = df['awards'].str.extract('Won \d+ (.*?)(?=\.)',expand=True)
df_awards.dropna(inplace=True)
df_awards.head(30)
df_awards[0].unique()
for award in df_awards[0].unique():
    df[award+' Won']=df['awards'].str.extract('Won (\d+) '+award,expand=True).fillna(0);
    df[award+' Nomination']=df['awards'].str.extract('Nominated for (\d+) '+award,expand=True).fillna(0);
    # values = {award+' Won': 0}
    # df.fillna(value=values)
df['Other wins']=df['awards'].str.extract('(\d+) win',expand=True).fillna(0);
df['Other nominations']=df['awards'].str.extract('(\d+) nomination',expand=True).fillna(0);
df.drop(columns =["awards"], inplace = True) 
# df.head(50)
# df['language'].value_counts()

# df.to_csv(r'/content/gdrive/My Drive/CSC 522 ALDA/Project/award_split.csv')
```
